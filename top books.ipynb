```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/bestsellers with categories.csv")
```

```python
df1=df
```

```python
df1.sort_values("User Rating",inplace=True)
```

```python
df1
```

```python
genre_type=df1.groupby("Genre")["Name"].count().reset_index()
```

# The Types of genre and their counts


```python
genre_type
```

```python

plt.figure(figsize=(10,10))
sns.barplot(x="Genre",y="Name",data=genre_type)
plt.show()
```

```python
top_5books_review=df1.sort_values("Reviews").head(5)
```

# top_5books_review

```python
top_5books_review
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x="Name",y="Reviews",data=top_5books_review)
plt.xticks(rotation=45)
plt.legend()


plt.show()
```

```python
top_5_author=df1.groupby("Author")["Name"].count().sort_values(ascending=False).head(5)
```

# top_5_author


```python
plt.figure(figsize=(10,10))
top_5_author.plot(kind="pie")
plt.legend()

plt.show()
```
